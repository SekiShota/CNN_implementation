```python
#モデル読み込み
import tensorflow.keras as keras
from keras.models import load_model

# model=load_model('./image1/cnndefault_ashtama.h5')
model=load_model('./image1/cnnarg-ashtama.h5')

model.summary()
```

```python
#テスト画像で検証
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

def detection(path):
    labels=["齋藤飛鳥", "玉森裕太"]
    
    for img in path:
        test_img=Image.open(img)
        test_img=test_img.convert("RGB")
        plt.axis('off')
        plt.imshow(test_img)
        plt.show()
        test_img=test_img.resize((64,64))

        x_test=np.asarray(test_img)
        x_test=x_test.reshape(-1, 64, 64, 3)

        pred=model.predict(x_test)[0]
        
        print(img)
        print(format(pred[0], '.10e'))
        if pred[0]<0.5: 
            print("予測結果：", labels[0])
        else:
            print("予測結果：", labels[1])
```

```python
import glob


#齋藤飛鳥の画像
asuka_path=glob.glob('./image1/asuka_test/*.jpg')

#玉森裕太の画像
tama_path=glob.glob('./image1/tama_test/*.jpg')
```

### 齋藤飛鳥の画像で検証

```python
detection(path=asuka_path)
```

### 類似度の確認

```python
#!/usr/bin/env python
# -*- coding: UTF-8 -*-

"""feature detection."""

import cv2
import os

def similar_level(target_img_path, comparing_files):
    target_img_path = target_img_path
    comparing_files = comparing_files
    IMG_SIZE = (200, 200)

    plt.imshow(Image.open(target_img_path))
    plt.show()

    target_img = cv2.imread(target_img_path, cv2.IMREAD_GRAYSCALE)
    target_img = cv2.resize(target_img, IMG_SIZE)

    bf = cv2.BFMatcher(cv2.NORM_HAMMING)
    detector = cv2.ORB_create()
#     detector = cv2.AKAZE_create()
    (target_kp, target_des) = detector.detectAndCompute(target_img, None)

    print('TARGET_FILE: %s' % (target_img_path))

    for file in comparing_files:
        plt.imshow(Image.open(file))
        plt.show()

        if file == '.DS_Store' or file == target_img_path:
            continue

        comparing_img_path = file
        try:
            comparing_img = cv2.imread(comparing_img_path, cv2.IMREAD_GRAYSCALE)
            comparing_img = cv2.resize(comparing_img, IMG_SIZE)
            (comparing_kp, comparing_des) = detector.detectAndCompute(comparing_img, None)
            matches = bf.match(target_des, comparing_des)
            dist = [m.distance for m in matches]
            ret = sum(dist) / len(dist)

        except cv2.error:
            ret = 100000

        print(file, ret)
```

```python
#CNN_defaultでの誤認識(AKAZE)

target_img_path = './image1/tama_test/2007.jpg'
comparing_files = glob.glob('./image1/asuka_test/*.jpg')

similar_level(target_img_path, comparing_files)
```
